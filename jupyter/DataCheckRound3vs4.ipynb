```python
# run PCA variance analysis on old data
# for iii in dfw.columns:
#    print(iii)
print("shape dfw: ",df3.shape)
# ASSUME protiens order starts with 'March 1' and ends with 'Zzz3'
col_start = 'March 1'
col_end = 'Zzz3'
icol_start = df3.columns.get_loc(col_start)
icol_end = df3.columns.get_loc(col_end)
print("icol_start: ",icol_start)
print("icol_end: ",icol_end)
feature_mat = df3.iloc[:,icol_start:icol_end+1]
print("shape feature_mat: ",feature_mat.shape)
pca3 = PCA(n_components=3)
components = pca3.fit_transform(feature_mat)
total_var3 = pca3.explained_variance_ratio_.sum() * 100
labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'}
print("total_var: ",total_var3)
```

```python
# run PCA variance analysis on new data
# for iii in dfw.columns:
#    print(iii)
print("shape dfw: ",df4.shape)
# ASSUME protiens order starts with 'March 1' and ends with 'Zzz3'
col_start = 'March1'
col_end = 'Zzz3'
icol_start = df4.columns.get_loc(col_start)
icol_end = df4.columns.get_loc(col_end)
print("icol_start: ",icol_start)
print("icol_end: ",icol_end)
feature_mat = df4.iloc[:,icol_start:icol_end+1]
print("shape feature_mat: ",feature_mat.shape)
pca4 = PCA(n_components=3)
components = pca4.fit_transform(feature_mat)
total_var4 = pca4.explained_variance_ratio_.sum() * 100
labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'}
print("total_var: ",total_var4)
```
